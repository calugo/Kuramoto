# FRUSTRATION AND NET TOPOLOGY

```python
import matplotlib.pyplot as plt
import sys 
sys.path.append('codes/')
import topologyselect as tps
from pygsl import rng as rn
import networkx as nx
from pygsl.const.m import *
#print pi
import initialstate as inist
import math

SEED=1
rk=rn.rng()
rk.set(SEED)

A=tps.graph(3,['fully']) #Fully connected
N=len(A.keys())

phi=inist.initialstate(N,rk,pi)

#axes.hist(phi.values())#
#axes.set_ylabel("Frequency($\phi_o$)",fontsize=40)
#axes.set_xlabel("Initial Phases $\phi_o$",fontsize=40)


G=nx.Graph()

for n in A.keys():
    for nk in A[n]:
        #print n, nk
        G.add_edge(n,nk)


plt.figure(figsize=(20,15))
plt.axis('off')
plt.axis('equal')
nx.draw_networkx(G,nx.circular_layout(G),node_size=2)


fig, axes = plt.subplots(1, 1, figsize=(15, 10))

for i in phi.keys():
    axes.plot(math.sin(phi[i]),math.cos(phi[i]),"o--")

circle1=plt.Circle((0,0),1.0,color='b',fill=False)
fig.gca().add_artist(circle1)
axes.set_title("Initial State",fontsize=40)
plt.axis("equal")
```
