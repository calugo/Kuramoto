First I present some results regarding the syncronisation time $\tau_{RS}$ which is the time it takes the system to achieve a syncronised state (What do you prefer, synchronisation, synchrinization British English or American English?) I use both without distinction). The system under study is the Kuramoto system on an undirected graph with adjacency matrix $\mathcal{A}_{ij}$:

$\dot{\phi}=\omega_{o,i}+\frac{\epsilon}{N}\sum_{j\neq i} \mathcal{A}_{ij} \sin(\phi_j-\phi_i)$

The syncronization times are for complete graphs composed by  $N=1000$ nodes and different  values of $\epsilon$, we  use $r(t)=\frac{1}{N}\sum_ie^{i\phi(t)}$ to determine  $\tau_{RS}$.

(Note 1. At the beginning I was intergrating $\dot{\phi}=\omega_{o,i}+\frac{\epsilon}{k_i}\sum_{j\neq i} \mathcal{A}_{ij} \sin(\phi_j-\phi_i)$ which gives different results but more boring)

# Temporal Network Dynamics 

As expected the transition to a syncronized state is faster and neater for large values of $\epsilon$. The question is which value should we employ as base for the temporal networks. Dunno if I mentioned before that the values $\omega_{i}$s are calculated at the beginning of the simulations as:
    
$\omega_{0,i}=\left[\mu+\eta\right] \mod (2\pi)$

with \mu=0.0 and $\eta$ a random variable centered in zero and unit variance. ARE YOU HAPPY WITH THIS?. The initial state is uniformily distributed in $\left[0,2 \pi \right]$

Having established the above then I am integrating the following scheme:

```python
SVG(filename='Kuramototnet.svg') 
```

The initial state is a fully connected graph which is integrated up to a time $t_1$, which can be $t_1<\tau_{SR}$ or $t_1>\tau_{SR}$. 

   Then the matrix $A(t)$ is re-wired by having a link between two nodes with probability $p$ or not with probability $1-p$. Once this has been done, then the system is integrated up to a time $t_1 + t_s$, and so forth.

I follow the evolution of $r(t)$ over time. I would like to know if you have any suggestion about which experiments I should carry on. So far I am only carrying out experiments with $\epsilon>=5$.
